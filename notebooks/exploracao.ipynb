```python
dados_csv = []

with open(path_csv, 'r') as file:
    spamreader = csv.DictReader(file, delimiter=',')
    for row in spamreader:
        dados_csv.append(row)
```

```python
nome_colunas_csv = list(dados_csv[0].keys())
nome_colunas_csv
```

```python
len(nome_colunas_csv)
```

```python
key_mapping = {'Nome do Item': 'Nome do Produto',
               'Classificação do Produto': 'Categoria do Produto',
               'Valor em Reais (R$)': 'Preço do Produto (R$)',
               'Quantidade em Estoque': 'Quantidade em Estoque',
               'Nome da Loja': 'Filial',
               'Data da Venda': 'Data da Venda',
                }
key_mapping
```

```python
new_dados_csv = []

for old_dict in dados_csv:
    dict_temp = {}
    for old_key, value in old_dict.items():
        dict_temp[key_mapping[old_key]] = value
    new_dados_csv.append(dict_temp)

new_dados_csv[0]
```

```python
combine_list = []
combine_list.extend(dados_json)
combine_list.extend(new_dados_csv)
```

```python
len(combine_list)
```

```python
nomes_colunas = list(combine_list[0].keys())
nomes_colunas
```

```python
path_dados_combinados = '../data_processed/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=nomes_colunas)
    writer.writeheader()

    for row in combine_list:
        writer.writerow(row)
```

```python
nomes_colunas = list(combine_list[-1].keys())
nomes_colunas
```

```python
dados_comninados_tabela = [nomes_colunas]

for row in combine_list:
    linha = []
    for coluna in nomes_colunas:
        linha.append(row.get(coluna, 'indisponível'))
    dados_comninados_tabela.append(linha)
```

```python
dados_comninados_tabela
```

```python
path_dados_combinados = '../data_processed/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.writer(file)
    writer.writerows(dados_comninados_tabela)
```
